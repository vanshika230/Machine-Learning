```python
import pandas as pd
```

```python
df=pd.read_csv('carprices.csv')
```

```python
df.head()
```

```python
dummies=pd.get_dummies(df['Car Model'])
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
final=merged.drop('Mercedez Benz C class',axis=1)
final
```

```python
X = final.drop(['Sell Price($)','Car Model'],axis='columns')
X
```

```python
y = final['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
model.predict([[45000,4,0,0]])
```
